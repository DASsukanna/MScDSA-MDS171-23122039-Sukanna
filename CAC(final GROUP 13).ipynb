```python
def routes():
    t = PrettyTable([Fore.YELLOW+'S.No', Fore.YELLOW+'Bus', Fore.YELLOW+'Route', Fore.YELLOW+'Timings']) #These 4 are the columns of the tables
    #To insert rows:
    t.add_row(['1.','Bus 301', 'Thicket Street - Portofino - Hill View - Clubview - MBA Block - Central Block', '7:30AM - 7:45AM - 8:05AM - 8:15AM - 8:22AM - 8:25AM'])
    t.add_row(['2.','Bus 302', 'Thicket Street - Portofino - Hill View - Clubview - MBA Block - Central Block', '12:30PM - 12:45PM - 1:05PM - 1:15PM - 1:22PM - 1:25PM'])
    t.add_row(['3.','Bus 303', 'Thicket Street - Portofino - Hill View - Clubview - MBA Block - Central Block', '3:40PM - 3:50PM - 4:05PM - 4:15PM - 4:20PM - 4:23PM'])
    t.add_row(['4.','Bus 304', 'Thicket Street - Portofino - Hill View - Clubview - MBA Block - Central Block', '7:30PM - 7:45PM - 8:05PM - 8:15PM - 8:22PM - 8:25PM'])
    print(t)
    print(Style.RESET_ALL)
```

```python
oldid=[]
passenger_details={}

def Passenger_details(bs_no):
    passengerid=str(random.randint(10000,99999))
    oldid.append(passengerid)
    unique(passengerid)
    storeDetails(passengerid,bs_no)
    print("Your passenger ID is: ",passengerid)

def unique(passengerid):
    #to make sure the Passenger ID generated is unique we add each ID generated to a set to avoid repitition
    #and change the id if it has been already generated. 
    #if the Passenger ID generated is unique the program continues.
    if passengerid in oldid:
        passengerid=str(random.randint(10000,99999))
        oldid.append(passengerid)
    else:
        oldid.append(passengerid)

def storeDetails(passengerid,bs_no):
    choice=input("Enter your details in order to proceed with the booking. If you wish to continue enter 'Y' else enter 'N'.").title()
    if choice=='Y':
        name=input("Enter the name of the student.").strip().title()
        reg=input("Enter the registration number of the student.")
        phone=input("Enter the mobile number of the student.")
        email=input("Enter the email id of the student.")
        passenger={
            "name":name,
            "reg. no":reg,
            "phone":phone,
            "email":email,
            "seat no.":bs_no
        }
        passenger_details[passengerid]=passenger

    else:
        print("You have chosen to exit.")
        exit()
    
```

```python
def view_bus(as_no):
    columns=2 #used to specify that the list is to be taken in pairs of two
    for first, second in zip(  #zip is used to comnine both the lists created below
        as_no[::columns] , as_no[1::columns]  #[::columns] creates a new list containing 1st, 3rd... elements, [1::columns] creates a new list containing 2nd, 4th... elements
    ):
        print(f'{Fore.GREEN}{first: <10}{Fore.GREEN}{second}') 
        print(Style.RESET_ALL)
```

```python
def booked(as_no,bs_no):
        columns=2 
        for first, second in zip(as_no[::columns] , as_no[1::columns]):
            if str(first) in bs_no:
                print(f'{Fore.RED}{first: <10}{Fore.GREEN}{second}')
                print(Style.RESET_ALL)
            elif str(second) in bs_no:
                print(f'{Fore.GREEN}{first: <10}{Fore.RED}{second}')
                print(Style.RESET_ALL)
            else:
                print(f'{Fore.GREEN}{first: <10}{Fore.GREEN}{second}')
                print(Style.RESET_ALL)
```

```python
def seat_no(bs_no):
        number=int(input("Enter the no of seats you want to book: "))
        for i in range(number):
            seat_number=input("Enter the seat number that you want to book: ")
            if seat_number in bs_no:
                print("Seat already booked by someone else, please try booking another seat.")
            else:
                bs_no.append(seat_number)
                pass
        bs_no.sort()
```

```python
def displayDetails(passenger):
    if passenger in passenger_details:
        print("Your bus pass is displayed below:")
        t = PrettyTable([ 'Name', 'Reg. No', 'Phone', 'Email ID']) 
        t.add_row([passenger_details[passenger]["name"], passenger_details[passenger]["reg. no"], passenger_details[passenger]["phone"], passenger_details[passenger]["email"]])
        print(t)
        print("Seats booked: ",passenger_details[passenger]["seat no."])
    else:
        print(Fore.RED+"Passenger ID does not exist.")
```

```python
def cancellation(passenger):
    if passenger in passenger_details:
        delete=[]
        bus_number=input("Enter the serial number of the bus you have booked a seat in.")
        if bus_number=='1':
            n=int(input("Enter the total number of seats you want to cancel."))
            for i in range(n):
                n1=input("Enter seat no. you want to delete:")
                delete.append(n1)
            for i in delete:
                if i in bs1:
                    bs1.remove(i)
            print("Your cancellation has been processed. The available seats are: ")
            booked(as1,bs1)
        elif bus_number=='2':
            n=int(input("Enter the total number of seats you want to cancel."))
            for i in range(n):
                n1=input("Enter seat no. you want to delete:")
                delete.append(n1)
            for i in delete:
                if i in bs2:
                    bs2.remove(i)
            print("Your cancellation has been processed. The available seats are: ")
            booked(as2,bs2)
        elif bus_number=='3':
            n=int(input("Enter the total number of seats you want to cancel."))
            for i in range(n):
                n1=input("Enter seat",i+1,"you want to delete:")
                delete.append(n1)
            for i in delete:
                if i in bs3:
                    bs3.remove(i)
            print("Your cancellation has been processed. The available seats are: ")
            booked(as3,bs3)
        elif bus_number=='4':
            n=int(input("Enter the total number of seats you want to cancel."))
            for i in range(n):
                n1=input("Enter seat",i+1,"you want to delete:")
                delete.append(n1)
            for i in delete:
                if i in bs4:
                    bs4.remove(i)
            print("Your cancellation has been processed. The available seats are: ")
            booked(as4,bs4)
        else:
            print(Fore.RED+"Invalid input")
    else:
        print(Fore.RED+"Passenger ID does not exist. Please try again.")
```
